```python
def EDA(df,y):
    import numpy as np
    import pandas as pd
    import matplotlib.pyplot as plt
    import seaborn as sns 
    
  
    
    columns = df.columns
    index = len(columns)
    count = 0
    heatmap = False
    displot = False
    histogram = False
    scatterplot = False
    barplot = False
    
   
   
    plt.figure(figsize=(8, 12))
    heatmap = sns.heatmap(df.corr()[[y]].sort_values(by=y, ascending=False), vmin=-1, vmax=1, annot=True, cmap='bwr')
    heatmap.set_title(f"Features Correlating with {y}", fontdict={'fontsize':18}, pad=16);
    plt.savefig(f"{y}_correlation.png") 
    desired_graph = input('What graph would you like to display?(displot,histogram,scatterplot,barplot):') 
    desired_graph = desired_graph.lower()
    
    
    while desired_graph != 'displot' or 'histogram' or 'scatterplot':
       

        if desired_graph == 'displot':
            displot = True
            break

        if desired_graph == 'histogram':
            histogram = True
            break

        if desired_graph == 'scatterplot':
            scatterplot = True
            break
        if desired_graph == 'barplot':
            scatterplot = True
            break
    
        
    
    while count < index:
        name = df.columns[count]
        
        if displot == True:
            sns.set_style('whitegrid')
            sns.distplot(df[name], kde = False, color ='red')
            plt.title(f"{name}values") 
            plt.show()
            plt.savefig(f"{y}_displot.png")
            
        if histogram == True:
            sns.set(style="darkgrid")
            sns.histplot(data=df, x= name)
            plt.axvline(df[name].mean(), label='mean',color='green')
            plt.axvline(df[name].median(), label='median',color='orange')
            plt.title(f"{name}values") 
            plt.legend()
            plt.show()
            plt.savefig(f"{y}_histogram.png")
            
        if scatterplot == True:
            sns.scatterplot(data = df, x = name, y = y)
            plt.title(f"{name}values")
            plt.show()
            plt.savefig(f"{y}_scatterplot.png")
        
        if barplot == True:
            sns.barplot(data=df, x=name, y=y)
            plt.title(f"{name}values") 
            plt.legend()
            plt.show()
            plt.savefig(f"{y}_barplot.png")
        
        
        count +=1
        
```
